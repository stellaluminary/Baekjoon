```python
def kaing_cha(T):
    for _ in range(T):
        M,N,x,y = map(int,input().split())
        
        sub_M_N=abs(M-N)
        bigger=M
        smaller=N
        circleNum=0
        M_bigger=True

        if M <= N:
            bigger=N
            smaller=M
            M_bigger=False

        tempsmaller = smaller

        for i in range(smaller):
            tempsmaller = smaller * (i+1)
            if tempsmaller % sub_M_N ==0:
                circleNum = tempsmaller // sub_M_N
                break
        countCir=0

        if M_bigger:
            s_x,s_y = x, x % smaller
            numbering = x

            for i in range(circleNum):
                if s_y == y:
                    print(numbering)
                    break      

                s_y=(s_y + sub_M_N) % smaller
                if s_y == 0:
                    s_y = smaller
                numbering += bigger
                countCir += 1

        else:
            s_x,s_y = y % smaller, y
            numbering=y

            for i in range(circleNum):
                if s_x == x:
                    print(numbering)
                    break

                s_x=(s_x + sub_M_N) % smaller
                numbering += bigger
                countCir += 1
                if s_x == 0:
                    s_x = smaller

        if countCir == circleNum:
            if s_x != M and s_y != N:
                print(-1)
                
T=int(input())
kaing_cha(T)
```

```python
def gcd(a, b):
    if a % b == 0:
        return b
    return gcd(b, a%b)

def lcm(a, b):
    return int(a*b/gcd(a,b))

t = int(input())

for _ in range(t):
    m, n, x, y = map(int, input().split())
    result = -1
    ma = lcm(m, n)
    if m == x and n != y:
        x = 0
    if m != x and n == y:
        y = 0
    if m == x and n == y:
        result = ma
    else:
        tmp = [x+m*i for i in range(ma//m+1) if x+m*i == ma or x+m*i % m == x]
        print(tmp)
        for e in tmp:
            if e % n == y and e % m == x:
                result = e
                break
    print(result)
```
