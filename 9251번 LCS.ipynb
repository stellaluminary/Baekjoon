# 제출 1 - 런타임 에러 : 긴 줄이 b가 되도록 설정해야함

```python
a = list(input())
b = list(input())
DP = [0] * (len(b)+1)
bf = [0] * (len(b)+1)
for i in range(len(a)):
    for k in range(i+1):
        for j in range(bf[k],len(b)):                    
            if a[i] == b[j]:                   
                if DP[i+1] <= DP[k]+1:
                    DP[i+1] = DP[k]+1                
                    bf[i+1] = j+1
                    break
                    
print(max(DP))
```

# 제출 2 - 틀렸습니다. 501B

```python
a = list(input())
b = list(input())
if len(a) > len(b):
    c = a
    a = b
    b = c
DP = [0] * (len(b)+1)
bf = [0] * (len(b)+1)
for i in range(len(a)):
    for k in range(i+1):
        if bf[k] >= len(b):
            continue
        for j in range(bf[k],len(b)):                    
            if a[i] == b[j]:                   
                if DP[i+1] <= DP[k]+1:
                    DP[i+1] = DP[k]+1                
                    bf[i+1] = j+1
                    break
print(max(DP))
```

# 제출 3 - 306B 37492KB 688ms - 정석 but 메모리 리소스가 큼

```python
a = list('AAABAABBAAAAA')
b = list('AAAAAAAAABBBBBBBB')

DP = [[0 for _ in range(len(a) + 1)] for _ in range(len(b) + 1)]

for i in range(len(a)):
    for j in range(len(b)):
        if a[i] == b[j]:
            DP[j+1][i+1] = DP[j][i] + 1
        else:
            DP[j+1][i+1] = max(DP[j][i+1], DP[j+1][i])
print(DP[-1][-1])
```
